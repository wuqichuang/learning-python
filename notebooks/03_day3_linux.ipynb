## 一、登录你的云服务器（10 分钟）

云服务器 = 一台远端的 Linux 电脑（一般是 Ubuntu），你通过 **SSH** 远程控制它。

在你自己电脑的终端（Windows 用 PowerShell / Mac 用"终端"）输入：

```bash
ssh 用户名@服务器IP        # 例如 ssh ubuntu@123.45.67.89
# 首次连接会问 yes/no → 输入 yes
# 然后输入密码（输入时屏幕不显示，正常现象，输完回车）
```

> 📌 看到提示符变成 `ubuntu@服务器名:~$` 就说明登录成功了。`~` 表示当前在自己的家目录。

## 三、权限与 sudo（30 分钟）

执行 `ls -l` 看到的这一串就是权限位：

```text
-rw-r--r-- 1 ubuntu ubuntu 12 Jul 17 10:00 hello.txt
└┬┘└┬┘└┬┘
 │   │   └─ 其他人(r 可读)
 │   └───── 同组人(r 可读)
 └───────── 所有者(rw 可读可写)
```

- `r`读=4，`w`写=2，`x`执行=1，三组数字相加 → `chmod 644 文件`
- `chmod +x script.sh`：给脚本加执行权限
- `sudo`：以管理员身份执行（装软件、改系统配置时用），如 `sudo apt update`

````python
!cd /tmp/linux_demo && touch script.sh && ls -l script.sh && chmod +x script.sh && ls -l script.sh
````

## 四、软件安装 apt（20 分钟，明天装 MySQL 就用它）

```bash
sudo apt update          # 刷新软件源列表（装东西前先跑这个）
sudo apt install -y 包名  # 安装，-y 自动确认
sudo apt remove 包名      # 卸载
```

示例：`sudo apt install -y tree`（装完后用 `tree` 命令可以树状显示目录）
