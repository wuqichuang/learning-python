## 一、Python 快速回顾（30 分钟）

只回顾后面每天都要用的四样东西：变量、列表、字典、循环。

````python
# 变量与基本类型
name = "李雷"          # 字符串 str
age = 22               # 整数 int
gpa = 3.75             # 浮点数 float
is_enrolled = True     # 布尔 bool
print(type(name), type(age), type(gpa), type(is_enrolled))
````

````python
# 列表 list：一组有顺序的数据
scores = [88, 92, 75, 60, 95]
print(scores[0])        # 第一个元素（下标从 0 开始）
print(scores[-1])       # 最后一个元素
print(len(scores))      # 长度
scores.append(83)       # 末尾添加
print("平均分:", sum(scores)/len(scores))
````

````python
# for 循环 + if 判断：批量处理的骨架
students = [
    {"name": "李雷",   "math": 58},
    {"name": "韩梅梅", "math": 91},
    {"name": "Jim",    "math": 76},
]
for s in students:
    if s["math"] >= 60:
        print(s["name"], "及格")
    else:
        print(s["name"], "【不及格】")
````

## 二、字符串处理（本日重点，30 分钟讲解）

### 2.1 切片：取出字符串的一部分

格式 `s[开始:结束:步长]`，**含头不含尾**。

````python
s = "Python数据处理"
print(s[0:6])    # 'Python' —— 取下标 0~5
print(s[6:])     # '数据处理' —— 从 6 到结尾
print(s[:6])     # 'Python' —— 从头到 5
print(s[-2:])    # '处理' —— 最后两个字符
print(s[::-1])   # 倒序
````

### 2.2 最常用的字符串方法

| 方法 | 作用 | 例子 |
|---|---|---|
| `s.strip()` | 去掉首尾空白/换行 | `" 张三 \n".strip()` → `"张三"` |
| `s.split(sep)` | 按分隔符拆成**列表** | `"a,b,c".split(",")` → `["a","b","c"]` |
| `sep.join(list)` | 把列表拼成一个字符串 | `",".join(["a","b"])` → `"a,b"` |
| `s.replace(old,new)` | 替换 | `"2026/01/01".replace("/","-")` |
| `s.find(sub)` | 查找位置，找不到返回 -1 | `"abc".find("b")` → `1` |
| `s.startswith()/endswith()` | 判断开头/结尾 | 常用来筛文件类型 |
| `s.upper()/lower()` | 大小写转换 | 统一英文数据格式 |
| `s.isdigit()/isalpha()` | 判断是否纯数字/纯字母 | 校验数据用 |

⚠️ 注意：字符串**不可变**，所有方法都返回**新字符串**，要赋值保存：`s = s.strip()`

````python
line = "  20260001,李雷,高三2班,88,92\n"
# 真实数据处理中最常见的组合拳：去空白 → 拆分 → 取字段
parts = line.strip().split(",")
print(parts)
sid, name, cls, math, english = parts
print(f"学号:{sid} 姓名:{name} 班级:{cls} 数学:{math} 英语:{english}")
````

### 2.3 f-string：把变量拼进文本（生成报告必用）

在字符串前加 `f`，用 `{}` 包住变量或表达式。

````python
name, math, english = "韩梅梅", 91, 84
# f-string 三种常用写法
print(f"{name} 的数学成绩是 {math} 分")
print(f"总分：{math + english}，平均分：{(math + english) / 2:.1f}")   # :.1f 保留 1 位小数
print(f"姓名：{name:<10}数学：{math:>5}")                              # :<10 左对齐占10格，:>5 右对齐占5格

# 批量生成报告行
rows = [("李雷", 58, 72), ("韩梅梅", 91, 84), ("Jim", 76, 88)]
print("\n" + "="*32)
for n, m, e in rows:
    print(f"| {n:<6} | 数学 {m:>3} | 英语 {e:>3} |")
````
