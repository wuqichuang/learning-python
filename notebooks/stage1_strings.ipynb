## Day 2 · 字符串基础：索引与切片

字符串可以看成一串字符排成的队伍，**每个字符都有编号（下标），从 0 开始**。
切片 `s[头:尾]` 取一段，**含头不含尾**——这是初学者最常栽跟头的地方。

````python
s = "Python"
print(s[0])     # 第 1 个字符：P
print(s[-1])    # 最后 1 个字符：n（负数从尾巴数）
print(s[1:4])   # 下标 1,2,3 → yth（含头不含尾）
print(s[:2])    # 从头到下标 1 → Py
print(s[2:])    # 下标 2 到结尾 → thon
print(s[::-1])  # 倒序 → nohtyP
````

### 字符串常用方法（重点，逐个敲一遍）

| 方法 | 作用 | 示例结果 |
|---|---|---|
| `strip()` | 去掉两端空白 | `"  a ".strip()` → `"a"` |
| `replace(旧, 新)` | 替换所有匹配 | `"hello".replace("l","L")` → `"heLLo"` |
| `split(分隔符)` | 拆成列表 | `"a,b".split(",")` → `["a","b"]` |
| `find(子串)` | 找子串位置，找不到返回 -1 | `"Python".find("th")` → `2` |
| `upper() / lower()` | 转大写 / 小写 | `"ab".upper()` → `"AB"` |
| `startswith() / endswith()` | 判断开头 / 结尾 | `"a.csv".endswith(".csv")` → `True` |
| `count(子串)` | 统计出现次数 | `"banana".count("a")` → `3` |

````python
text = "  Data-Science-2026  "
print(text.strip())          # 去掉两端空白
print(text.strip().lower())  # 方法可以连着用：先去空白再转小写

csv_line = "张三,88,二班"
parts = csv_line.split(",")  # 拆成列表
print(parts)
print("姓名是：", parts[0])

print("hello".replace("l", "L"))
print("Python".find("th"))
print("score.csv".endswith(".csv"))
print("banana".count("a"))
````

## Day 3 · 字符串拼接与格式化

把变量和文字组合成完整句子，有三种常见方式，**推荐优先用 f-string**：

| 方式 | 写法 | 适用场景 |
|---|---|---|
| `+` 号 | `"你好" + name` | 少量拼接，直观 |
| `join()` | `"-".join(列表)` | 拼接列表，效率最高 |
| f-string | `f"你好{name}"` | 变量嵌入文本，最灵活 |

````python
# 方式一：+ 号（少量拼接时直观）
first = "数据"
last = "处理"
print(first + " " + last)

# 方式二：join()（拼接整个列表）
parts = ["2026", "07", "17"]
print("-".join(parts))

# 方式三：f-string（大括号里直接放变量或表达式）
name = "小李"
score = 95.456
print(f"{name}的成绩是{score}分")
print(f"{name}的成绩是{score:.1f}分")   # :.1f 表示保留 1 位小数
````

````python
# f-string 常用格式控制
print(f"{3.14159:.2f}")     # 保留 2 位小数 → 3.14
print(f"{7:03d}")           # 整数不足 3 位前面补 0 → 007
print(f"{'语文':>8}")       # 右对齐，占 8 个字符宽度
print(f"{'数学':<8}|")      # 左对齐，占 8 个字符宽度

# 大括号里还可以写表达式
math, english = 88.5, 92.0
print(f"平均分：{(math + english) / 2:.1f}")
````
